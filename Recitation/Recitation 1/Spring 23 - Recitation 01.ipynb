```python
# Comparing numpy vectorization vs loops

import time

# lists
list_a = [i for i in range(1000000)]
list_b = [j for j in range(10, 1000010)]

# lists in array form
array_a = np.array(list_a)
array_b = np.array(list_b)

# form loop over lists
start = time.time()
total = 0
for i in range(len(list_a)):
    total += list_a[i] * list_b[i]
list_time = time.time() - start
print(total)

print('*'*100)

# vectorization multiply
start = time.time()
# array_a = np.array(list_a)
# array_b = np.array(list_b)
total = np.dot(array_a, array_b)
np_time = time.time() - start
print(total)
print('Speed up:', list_time / np_time)
```

```python
# Task: plot the function of y = sin(x)   [input: [-pi, pi], 50]

# data generation
x = np.linspace(start=-np.pi, stop=np.pi, num=50)
y = np.sin(x)
# axis
plt.axis([-np.pi, np.pi, -1., 1.])

# xticks
#plt.xticks(ticks=[-np.pi, 0, np.pi], labels=['\pi', 0, 'pi'])

plt.xticks(ticks=[-np.pi, 0, np.pi], labels=[r'$-\pi$', 0, r'$\pi$'])

# xlabel, ylabel
plt.xlabel('x')
plt.ylabel(r'$f=\sin(x)$')
# title
plt.title(r'$f=\sin(x)$')
# mathematical expression (Latex): https://matplotlib.org/3.1.1/tutorials/text/mathtext.html

plt.plot(x, y)
plt.show()
```
